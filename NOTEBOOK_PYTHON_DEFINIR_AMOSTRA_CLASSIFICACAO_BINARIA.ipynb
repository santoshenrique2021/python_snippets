```python
#Imports
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#Base de treino
df = pd.read_csv('df_treino_m6.csv', sep = ';')
#Visualização
df.head(2)
```

```python
# 3. Amostragem estratificada (5.000 observações)
n_amostra = 5000
amostra, _ = train_test_split(
    df,
    train_size=n_amostra,
    stratify=df['ALVO'],  # Mantém a proporção do ALVO
    random_state=42
)
```

```python
amostra.head(3)
```

```python
# 5. Resultado (opcional: salvar em CSV)
amostra.to_csv('amostra_estratificada.csv', index=False)
```
